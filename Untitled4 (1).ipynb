```python
import pandas as pd
import numpy as np
from xgboost import XGBClassifier
file=open("E://数据挖掘练习//客户逾期预测//after_RF_train.csv",encoding="utf_8_sig")
train_data=pd.read_csv(file)
train_data=train_data.dropna()
x_train=train_data.iloc[:,1:]
y_train=train_data.iloc[:,0]

file1=open("E://数据挖掘练习//客户逾期预测//new_test_data.csv",encoding='utf_8_sig')
test_data=pd.read_csv(file1)
test_data=test_data.dropna()
new_test_data=test_data

for i in range(test_data.shape[1]):
    if test_data.columns[i] not in train_data.columns:
        new_test_data=new_test_data.drop(test_data.columns[i],axis=1)
    else:
        continue
x_test=new_test_data.iloc[:,1:]
y_test=new_test_data.iloc[:,0]
y_test=y_test.reset_index(drop=True)

xgb=XGBClassifier(n_estimators=1500)
xgb.fit(x_train,y_train)
y_predict=xgb.predict(x_test)
k=0
for i in range(len(y_predict)):
    if y_test[i]==y_predict[i]:
        k=k+1
print(k/len(y_predict))
```

```python
from sklearn.metrics import confusion_matrix
c=confusion_matrix(y_test,y_predict)
print(c)
```

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt
plot_importance(xgb)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=3000)
rf.fit(x_train,y_train)
y_predict=rf.predict(x_test)
k=0
for i in range(len(y_predict)):
    if y_test[i]==y_predict[i]:
        k=k+1
print(k/len(y_predict))
c=confusion_matrix(y_test,y_predict)
print(c)
```
